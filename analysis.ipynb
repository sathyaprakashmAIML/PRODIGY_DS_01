```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Chinas Population En.csv")
df.head()
```

```python
df.columns
```

```python
plt.title("Distribution")
sns.histplot(df['Unnamed: 0'],bins=20,kde=True)
plt.show()
```

```python
plt.title("Barplot for gender")
sns.barplot(x=df['The male population (ten thousand people)'],y=df["At the end of the total population (ten thousand people)"],data=df)
plt.show()
```

```python
df['Urban population (ten thousand people)'].plot(kind='line', figsize=(10, 6))
plt.title('Trend of Urban Population')
plt.ylabel('Urban Population (ten thousand people)')
plt.xlabel('Index')
plt.show()
```

```python
sns.boxplot(data=df[['The male population (ten thousand people)', 'Female population (ten thousand people)']])
plt.title('Distribution of Male vs Female Population')
plt.xticks([0, 1], ['Male', 'Female'])
plt.show()
```

```python
df['Rural population (ten thousand people)'].plot(kind='line', figsize=(10, 6), color='orange')
plt.title('Trend of Rural Population')
plt.ylabel('Rural Population (ten thousand people)')
plt.xlabel('Index')
plt.show()
```
